```python
def pages_within_restaurant(base_url, pages):
    # List to store the URLs
    url_list = []

    # Loop to generate URLs
    for i in range(0, pages, 10):
        full_url = f"{base_url}?start={i}"
        url_list.append(full_url)
        print(url_list)

    return url_list
```

```python
all_restaurants_url_lists = []
Pages_single_restaurtant_url_list =[]



root_url = "https://www.yelp.ca"

# set a smaller value due to API limitations, can be increased ( number of pages of reviews for each restaurant to  limit number of pages per restaurant )
pages = 40

for url in Restaurant_url_list:
    full_url = root_url + url
    r = requests.get(full_url, headers={"Authorization": "Bearer %s" % YELP_TOKEN})
    print(f"Status code for {full_url}: {r.status_code}")
    # fn call for within each restaurant
    Pages_single_restaurtant_url_list = pages_within_restaurant(full_url, pages)


    # Append the list of URLs for this restaurant to the main list
    # this list contains each restaurant and all the pages within it
    all_restaurants_url_lists.append(Pages_single_restaurtant_url_list)


```

The scape_single_page function contains three parts:


1.   We scrape to the number of stars in each page (  out of 5)
2.   We scrape each review within a page

We add it to a dictionary (can also be a tuple list)


1.   since we are doing a classification problem, we only positive or negative
so I set up a classification column if the rating is >2.5 to be positive


Then we count the total number of positive reviews (and limit it at 500) to create a balanced dataset (See the next textbox for details)







```python
def scrape_single_page(reviews, positive_count, max_positive_count):
    scraped_data = []
    #total_count = 0

    for index, review in enumerate(reviews, start=0):
        dic = {}

        # Check if the elements exist before attempting to access them
        rating_element = review.find('span', attrs={'class': "css-1d8srnw"})
        content_element = review.find('span', attrs={'class': "raw__09f24__T4Ezm"})
        #total_count += 1

        if rating_element and content_element:
            rating = rating_element.find('div').get('aria-label')
            rating = float(re.findall('\d+', rating)[0])
            review_content = content_element.get_text().strip()

            #dic['rating'] = rating
            dic['content'] = review_content

            sentiment = 'positive' if rating > 2.5 else 'negative'
            dic['sentiment'] = sentiment

            if sentiment == 'positive':
                positive_count += 1
                if positive_count > max_positive_count:
                    print(f"Skipping positive review {index} to balance the dataset")
                    continue  # Skip this positive review and move to the next one

            scraped_data.append(dic)
        else:
            print(f"Skipping review {index} due to missing elements")

    return scraped_data, positive_count

```

```python
all_scraped_data = []
positive_count = 0
max_positive_count = 500  # Adjust as needed

for url_list in all_restaurants_url_lists:
    for index, url in enumerate(url_list, 1):
        r = requests.get(url, headers={"Authorization": "Bearer %s" % YELP_TOKEN})
        text = BeautifulSoup(r.text, 'html.parser')
        reviews = text.find_all('li', attrs={'class': "css-1q2nwpv"})

        scraped_data, positive_count = scrape_single_page(reviews, positive_count, max_positive_count)
        all_scraped_data.extend(scraped_data)

        time.sleep(random.randint(1, 3))
        print(f'Finished processing {url}')
        print(f'Finished page {index}')


# Write the list of dictionaries to a JSON file
json_file_path = 'reviews.json'
with open(json_file_path, 'w', encoding='utf-8') as jsonfile:
    json.dump(all_scraped_data, jsonfile, ensure_ascii=False, indent=2)
```
